```python
%matplotlib notebook
from pylab import *
import numpy as np
#from mpl_toolkits.mplot3d import Axes3D
```

```python
r = 12.87
winkel = np.linspace(-np.pi, np.pi, 100)
x = np.cos(winkel)*r
y = np.sin(winkel)*r
```

```python
plt.plot(x,y, '--r')
plt.title('Roter gestrichelter Kreis')
plt.xlabel(r'$\cos(t)$')
plt.ylabel(r'$\sin(t)$')
plt.axis('equal')
#plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.savefig('kreis.pdf')
plt.show()
```
