```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
def plot_cylinder(ax3d,x,y,z,r=1,h=1,Nphi=20,Nh=2,fc=[1,0.5,0,0.2]) :
    PHI,H = np.meshgrid(linspace(0,2*pi,Nphi),linspace(0,h,Nh))
    X=x+r*cos(PHI)
    Y=y+r*sin(PHI)
    c=ax3d.plot_surface(X,Y,z+H,rstride=1,cstride=1)
    c.set_facecolor(fc)
    return c

def plot_cone(ax3d,x,y,z,r=1,h=1,Nphi=20,Nr=2,fc=[0,1,0,0.2]) :
    PHI,R = np.meshgrid(linspace(0,2*pi,Nphi),linspace(0,r,Nr))
    X=x+R*cos(PHI)
    Y=y+R*sin(PHI)
    Z=z+(r-R)/r*h
    c=ax3d.plot_surface(X,Y,Z,rstride=1,cstride=1)
    c.set_facecolor(fc)
    return c

def plot_snowflakes(ax3d,Ix,Iy,Iz,N) :
    x=Ix[0]+np.random.rand(N)*Ix[1]
    y=Iy[0]+np.random.rand(N)*Iy[1]
    z=Iz[0]+np.random.rand(N)*Iz[1]
    ax3d.scatter(x,y,z,marker='*',s=50,c=[1,1,0,0.5])
```

```python
fig=figure('3DPlot')
ax3d = fig.add_subplot(111,projection='3d')
plot_cylinder(ax3d,0,0,0,r=1,h=28)
plot_cone(ax3d,0,0,26,r=2,h=4)
plot_cone(ax3d,0,0,18,r=4,h=10)
plot_cone(ax3d,0,0,2,r=8,h=20)
plot_snowflakes(ax3d,[-8,16],[-8,16],[0,30],N=100)
#ax3d.set_aspect('equal')
show()
```
