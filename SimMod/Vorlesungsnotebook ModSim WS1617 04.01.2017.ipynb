```python
%matplotlib notebook
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
L=100
p=np.linspace(-1,1,L)
q=np.linspace(0,2*np.pi,L)
P,Q=np.meshgrid(p,q)
#R=P**0.2+0.2*sin(P*2*pi)
X=(1+P/2*cos(Q/2))*cos(Q)
Y=(1+P/2*cos(Q/2))*sin(Q)
Z=(P/2*cos(Q/2))
F=np.zeros((L,L,3))
F[:,:,0]=(1+sin(Q))/2
F[:,:,1]=(1+cos(Q))/2
F[:,:,2]=1/2
```

```python
fig=figure('3DPlot')
ax = fig.add_subplot(111,projection='3d')

#x.plot_surface.
ax.plot_surface(X,Y,Z,rstride=2,cstride=2,facecolors=F,alpha=0.4)
show()
```

Klein-Bottle

```python
L=100
p=np.linspace(0,2*pi,L)
q=np.linspace(0,2*pi,L)
P,Q=np.meshgrid(p,q)

R=4*(1-cos(Q)/2)
X=6*cos(Q)*(1+sin(Q))+R*cos(Q)*cos(P)
X[Q>=pi] = (6*cos(Q)*(1+sin(Q))+R*cos(P+pi))[Q>=pi]

Y=16*sin(Q)+R*sin(Q)*cos(Q)
Y[Q>=pi]=16*sin(Q[Q>=pi])


Z=R*sin(P)

F=np.zeros((L,L,3))
F[:,:,0]=(1+sin(Q))/2
F[:,:,1]=(1+cos(Q))/2
F[:,:,2]=1/2
```

```python
fig=figure('3DPlot')
ax = fig.add_subplot(111,projection='3d')

#x.plot_surface.
ax.plot_surface(X,Y,Z,rstride=2,cstride=2,facecolors=F,alpha=0.4)
show()
```
