```python
import numpy as np
s2=0.1
m=0.01

Nn = np.array([1,10,50,100,500,1000])
Nmax=max(Nn)

# schreiben einer datei, die die positionen x von 200 Molekuelen zur zeit n beschreiben 

f = open('single_particle_random_walks.dat','w')
f.write('#200 single particle positions at 6 different time steps\n')
f.write('# nr. x(n) for n=')
for n in Nn :
    f.write(str(n)+" ")
f.write("\n#\n")

for nr in range(200) :
    f.write(str(nr)+" ")
    x = np.cumsum( (np.random.rand(Nmax)-0.5) * np.sqrt(s2) + m )
    for n in Nn :
        f.write("{:10.5f} ".format(x[n-1]))
    f.write("\n")
    
f.close()
```

```python
%matplotlib inline
from pylab import *

# datei laden
data = loadtxt("single_particle_random_walks.dat")

# zeitintervall / x-achse
n = [1,10,50,100,500,1000]

# definieren einer neuen matrix aus der tabelle ohne die 1. spalte
xn = ones((200,6))
for i in range(6):
    xn[:,i] = data[:,i+1]
    
# plotten der 200 zeilen    
for y in range(200):
    plot(n,xn[y,:],color=(0,0,0,0.5))

# plot beschriftung    
xticks(n)
xlabel('$\ n$')
ylabel('$\ x(n)$')
xlim(-100,1200)
show()
```

```python
# histogramm

# fuer n = 1

subplot(911)
bins1 = linspace(min(xn[:,0]),max(xn[:,0]),20)
hist(xn[:,0],bins1)
show()

# fuer n= 10

subplot(912)
bins2 = linspace(min(xn[:,1]),max(xn[:,1]),20)
hist(xn[:,1],bins2)
show()

# fuer n=50

subplot(913)
bins3 = linspace(min(xn[:,2]),max(xn[:,2]),20)
hist(xn[:,2],bins3)
show()

# fuer n=100
subplot(914)
bins4 = linspace(min(xn[:,3]),max(xn[:,3]),20)
hist(xn[:,3],bins4)
show()

# fuer n=500
subplot(915)
bins5 = linspace(min(xn[:,4]),max(xn[:,4]),20)
hist(xn[:,4],bins5)
show()

# fuer n=1000
subplot(916)
bins6 = linspace(min(xn[:,5]),max(xn[:,5]),20)
hist(xn[:,5],bins6)
show()
```

```python
# mittelwert und standardabweichung bestimmen

mx = []
stdx = []
x = [1,2,3,4,5,6]

for m in range(6):
    mx.append(mean(xn[:,m]))
    
for s in range(6):
    stdx.append(std(xn[:,s]))
    
```

```python
# errorbar
for e in range(6):
    errorbar(x[e],mx[e],stdx[e])
    
xticks(x,[1,2,3,4,5,6])
xlabel('$\ n$')
ylabel('$\ mittelwerte von x(n)$')
show()
```
